```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

```

```python
# Load the dataset
df = pd.read_csv("../data/mobile_data_new.csv")  # use relative path if needed

# View basic info
print("ðŸ”¹ Dataset Shape:", df.shape)
print("\nðŸ”¹ Data Types:\n", df.dtypes)

```

```python
# Check for missing values
print("\nðŸ”¹ Missing Values:\n", df.isnull().sum())

# View class distribution
print("\nðŸ”¹ Target Distribution:\n", df['price_range'].value_counts(normalize=True))
```

```python
# Correlation heatmap
plt.figure(figsize=(14,10))
sns.heatmap(df.corr(), annot=True, cmap="coolwarm", linewidths=0.5,fmt=".2f")
plt.title("Feature Correlation Heatmap")
plt.show()
```

```python
# Distribution of a few features
for col in ['battery_power', 'ram', 'int_memory', 'px_height']:
    plt.figure(figsize=(6, 4))
    sns.histplot(data=df, x=col, hue='price_range', kde=True, palette="Set2")
    plt.title(f'Distribution of {col} by Price Range')
    plt.show()
```
